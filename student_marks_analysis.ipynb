```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Student": ["A","B","C","D","E"],
    "Maths": [80,70,60,90,85],
    "Science": [75,65,55,95,80]
}

df = pd.DataFrame(data)

df["Average"] = (df["Maths"] + df["Science"]) / 2

print(df)

df.plot(x="Student", y=["Maths","Science"], kind="bar")
plt.title("Student Marks Analysis")
plt.show()
```
